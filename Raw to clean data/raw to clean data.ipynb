# .iloc[]:

### It stands for "integer location" and is used to select data based on integer position.
### You use it to select rows and columns by their index positions (i.e., integer-based indexing).
### Example: df.iloc[0, 1] will select the element in the first row and second column of the DataFrame.

```python
data.iloc[:, 1] 
```

```python
data['ADDRES'].str.split(r'[, ]',expand=True) 
```

```python
# after split we are storing in a different column name(cross','sector', 'state', 'zipcode')
data[['cross','sector', 'state', 'zipcode']]=data['ADDRES'].str.split(r'[ ]',expand=True)
data.head()
```

```python

data['ADDRES'].str.split(r'[, ]',expand=True)
```

```python
# Above we can see the ADDRES data splitted into 5 but need only 0,2,4,5 columns so here we are using .iloc to get which data we need 
data[['cross','sector', 'state', 'zipcode']]=data['ADDRES'].str.split(r'[, ]',expand=True).iloc[:,[0,2,4,5]]
```

```python
data.head()
```

```python
data[['day','month','year']]=data['date'].str.split('[ ]',expand=True)
```

```python
data
```

```python
data['time'].str.split('[ ]',expand=True)
```

```python
data[['hours','mins','sec']]=data['time'].str.split('[ ]',expand=True).iloc[:,[0,2,4]]
```

```python
data
```

```python
# we are creating new variable to create new csv file and just use which data we need
clean_data=data[["cross","sector","state","zipcode","day","month",'year','hours','mins','sec']]
```

```python
clean_data
```

```python
# we are saving the data using .to_CSV ("file name ")
clean_data.to_csv('clean data.csv')
```
