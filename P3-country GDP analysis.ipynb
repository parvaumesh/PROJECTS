### visualizations


```python
import matplotlib.pyplot as plt
import seaborn as sns  #seaborn is to create attractive and informative statistical graphics
%matplotlib inline
plt.rcParams['figure.figsize']= 10,5 # output graph size
import warnings
warnings.filterwarnings('ignore')
```

```python
df['InternetUsers']
```

```python
vis1=sns.distplot(df['InternetUsers'])
plt.show(vis1)
```

```python
vis2=sns.displot(df['InternetUsers'])
```

```python
vis3=sns.distplot(df['InternetUsers'], bins=10)
```

```python
vis4 = sns.boxplot(data=df,x="IncomeGroup",y='BirthRate')
```

```python
vis3 = sns.lmplot(data =df,x = 'InternetUsers', y = 'BirthRate', fit_reg = False) #lm - linear model
```

```python
vis4 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate')
```

```python
vis5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup') #hue - parameter for color
```

```python
vis5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup', height = 10)
```
